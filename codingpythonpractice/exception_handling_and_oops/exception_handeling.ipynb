```python
try:
    a=5
    b=int(input("please enter the value for b"))
    result=a/b
except ValueError as ve:
    print(ve)
except ZeroDivisionError as ze:
    print(ze)
except Exception as ex:
    print(ex)

```

```python
# what if i wnat to print the result here we use else operator
try:
    a=5
    b=int(input("please enter the value for b"))
    result=a/b
except ValueError as ve:
    print(ve)
except ZeroDivisionError as ze:
    print(ze)
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
```

```python
#finlly is used to execute the program despite the error used in critical situation when 

try:
    a=5
    b=int(input("please enter the value for b"))
    result=a/b
except ValueError as ve:
    print(ve)
except ZeroDivisionError as ze:
    print(ze)
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
finally:
    print('execution of the code is complete')
# this completely executes the code even when exceptions are triggred
# this massively helps in data base connections
```
