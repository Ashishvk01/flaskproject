Here we will create a file whos name is example.txt then what we will try to open it using read mode 

```python
with open('example.txt','r') as file :
    reader=file.read()

print(reader)
```

```python
#now here is how to do write operation inside the file 
with open('example.txt','w') as file:
    file.write('\n this will completely override the file that we have created')

#to avoide the problem we are going to use a file mode a
with open('example.txt','a') as file :
    file.write('this will not override but will start in the same line as of previous file so remember to leave a space at last ')

with open('example.txt','a') as file :
    file.write('\n')

#if we use /n 
with open('example.txt','a') as file :
    file.write(' if we use slash n in previous file it we start in the bext line')
```

```python
#here is the better was to do it, if you want to write multiple line in a file 

# create a list like this  
lines=['hi this is avk\n','i want to explain the multiple lines of code to myself \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
#what about the binary file that have .bin with them

#remember binary file should be written in this way only
data=b'\x00\x01\x02\x03'

with open('example.bin','wb') as file :
    file.write(data)


#how to read them
with open('example.bin','rb') as file :
    reader=file.read()

print(reader)
```

```python
# how to read the content of the file and transfer it to another file 
with open('example.txt','r') as file:
    content=file.read()

with open('example1.txt','w') as file:
    file.write(content)
```

```python
# when to use w+ 

with open('dext.txt','w+') as file:
    file.write("hgihiehqih iiefhi gefb \n")
    file.write("ksdk eqgfiiw igidfk igqekfi qegfig \n")

# when you do this use seek function to move cursor to beging
    file.seek(0)

# now you can read the file content very easily
with open('dext.txt','r+') as file :
    content=file.read()

print(content)
```

working with file path

```python
# we need to create a directory:
import os


new_directory='package2'
os.mkdir(new_directory)
print(f"new directory '{new_directory}' created")
```

```python
#listing the files and directory

path=os.listdir('.')
print(path)
```

```python
#os.path.join() is used to combine directory and file names into a valid file path in a platform-independent way.
folder='package'
file='dext.txt'
path=os.path.join(os.getcwd(),folder,file)
print(path)


```

```python
#how to check if file exist or not
if os.path.exists('example.txt'):
    print("yes it exist")

else:
    print('path does not exist')
```

```python
#also print the path if it exist
if os.path.exists('example.txt'):
    # print(os.path.join(os.getcwd(),'example.txt')) #this is exact use case of path.join
      print(os.path.abspath('example.txt'))

else:
    print('path does not exist')
```

```python
#check weather it is file or directory present in this folder or tell it is niether 
entity='package.txt'
if os.path.isfile(entity):
    print('it is a file')
elif os.path.isdir(entity):
    print('it is a directory')
else:
    print('this is nothing')
```
