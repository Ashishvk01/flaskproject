Tupels:these are ordered collection of items that are immuteable. They are similar to list but the lists are mutable.


```python
#how to create a tuple 
tpl=()
tpl=(1,2,'guy',False,33.33)
print(tpl)
```

```python
#how tuple index is used how slicing works is pretty similar to list
numbers=(1,2,3,4,56)
print(numbers[0])
print(numbers[0:3])
print(numbers[::-1])
print(numbers[::2])
print(numbers[::-2])
```

```python
# a list can be coverted into tpl and vice-versa
numbers=[1,2,3,4,5]
tpl=tuple(numbers)
print(type(tpl))
print(tpl)

tpl=list(numbers)
print(type(tpl))
print(tpl)
```

```python
# tuple operations
# '+' will add concatenate the tuple
numbers=(1,2,3,5)
colours=('red','blue','green','pink','yellow')
concatenate=numbers+colours
print(concatenate)
# '*' will repeat the tuple the number of times it is multiplied
numbers=numbers*3
print(numbers)
# regular list operation
print(numbers.count(1)) #it counts how many number of 1 are there with in tuple 
print(numbers.index(2)) #it give index of 2 were it is first present int list
```

```python
# how list are mutable and tuple is not
numbers=[1,2,3,4,5]
numbers[1]='ash'
print(numbers)
# list can change the content even though they are assigned

numbers=(1,2,3,4,5)
numbers[1]='ash'
# this cannot assign again when it already assigned
```

```python
#how to pack and unpack a tuple
packed_tuple=1,'ftht',False
print(packed_tuple)
a,b,c=packed_tuple
print(a)
print(b)
print(c)

#how to unpack with the help of a star
numbers=[1,2,3,4,5,6]
a,*b,c=numbers
print(a)
print(*b)
print(c)
```

```python
# nested tuples

tpl=((1,2,3),(4,5,6),(7,8,9))
print(tpl)
print(tpl[0])
print(tpl[0][1])
print(tpl[0][::-1])
print(tpl[0][0:2])
```

```python
# also you can add tuple in the list
list=[[1,2,3],[4,5,6],(7,8,9)]
print(list[2][2])
# we can add tuple inside the list
tpl=((1,2,3),(2,3,4),[1,5,9])
print(tpl[2][2])
```
