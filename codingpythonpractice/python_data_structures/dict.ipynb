dictonary: these are collection of items which are unordered and stored in key-value pairs .keys are immutable and should be unique, but values are mutable .

```python
#how to use a dictonary
ict={'name':'ashish','age':24,'gender_male':True}
print(ict)

# how to axcess the list
print(ict['name'])

# how to axcess it using get 
print(ict.get('gender_male'))

#what is some one wants to axcess but its in the list
print(ict.get('gender_female'))

#what is some one wants to axcess but its in the list but we want to give default value for it
print(ict.get('gender_female','not_avail'))
```

```python
# why keys should be unique
student={'name':'ash','age':24,'name':24}
print(student)
# when you see the result it takes the key which is most resent and the previous ones are lost.
```

```python
# how to do the addition , modification , deletion in dictonary
student1={'name':'ash','age':24,'grade':'A'}
student1['age']=23 #modification
student1['address']='third cross ghevarchand nagar' #addition
print(student1)
del student1['age'] #use del key-word to delete
print(student1)
```

```python
student1={'name':'ash','age':24,'grade':'A'}
keys=student1.keys()# keys values items functions 
values=student1.values()
pairs=student1.items()
print(keys)
print(values)
print(pairs)
```

```python
# when transfer values of one data_structures to another it 
student2=student1
print(student2)
print(student1)

student1['name']='hsa'
print(student2)
print(student1)
# even if we make changes individually both of the dict will get affected 
# so to avoid that we can use shallow copy by using copy fuction

student3=student2.copy()
print(student2)
print(student3)

student2['age']=21
print(student2)
print(student3)

```

```python
students={
    'student1':{'name':'ash','age':24,'grade':'A'},
    'student2':{'name':'kas','age':24,'grade':'B'},
    'student3':{'name':'sam','age':24,'grade':'c'}
}


#how to iterate over these dictonary
for students_id,students_info in students.items():
    print(f"{students_id},{students_info}")
    for key,values in students_info.items():
        print(f"{key},{values}")
```

```python
num={x:x**2 for x in range(5) }
print(num)

num={x:x**2 for x in range(5) if x%2==0 }
print(num)


#use a dictonary to count the frequency of elements in the list
lst=[1,2,3,1,2,3,4,5,6,4,5,6]
dict={x:lst.count(x) for x in lst}
print(dict)
# dict will always have unique value as keys so they are not repeated
```
