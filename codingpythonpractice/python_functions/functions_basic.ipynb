FUNCTIONS: The functions are the block of code. This helps to organize ,improve readability and reuse the code.

```python
#HOW TO CREATE A SIMPLE FUNCTION
def even_or_odd(num):
    """this is a function to tell weather a number is even or odd"""
    if num%2==0:
        print("even")
    else:
        print('odd') 
    
    
```

```python
#how to call the function
even_or_odd(15)

#or
print('please enter the number that you want to check even or odd')
num=int(input('please enter the number that you want to check even or odd'))
even_or_odd(num)

```

```python
#function also takes multiple inputs
def addition(a,b):
    sum=a+b
    print(sum)

addition(3,5)
```

```python
# A function can have default value
def wel(name='guest'):
    print(f"welcome {name} this our house")

wel()

# which you can change while calling the function
wel('john')
```

```python
#positional arguments: used when we have veriable length arguments
#ex 1
def print_anything(*args):   #usually used with *args *is mandotory args can be any keyword
    print(*args)

print_anything(1,2,3,4,5,'ash')

#ex 2
def print_anything1(*args):
    for numbers in args:
        print(numbers)

print_anything1(1,2,3,4,5,'ash')
```

```python
# keyword argument :used when we have veriable length keywords
def print_anything2(**kwargs):
    for key,val in kwargs.items():  #whenever you use a key-value pair use .items()
        print(f"{key}:{val}")

print_anything2(name='ash',age=24)


# when you wnat to use both pos arg and keyword arg tougether 
def print_anything3(*args,**kwargs,):  #cannot change the order
    for numbers in args:
        print(numbers)
    for key,val in kwargs.items():  
        print(f"{key}:{val}")


print_anything3(1,2,3,4,5,name='ash',age='post',grade='a')

# print_anything3(name='ash',age='post',grade='a',1,2,3,4,5) cannot change the order

# reason: "Once you start naming things, you are not allowed to give unnamed things anymore."

 
```
