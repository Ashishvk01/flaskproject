A lambda function is an anonymous, one-line function in Python used for short 
and simple operations without using the `def` keyword.

```python
#how to use a normal function
def addition(a,b):
    return a+b

addition(2,4)
```

```python
#how to do it with lamda function
addition1=lambda a,b:a+b
print(type(addition1))
addition1(9,0)
```

```python
#lets do some more example 
even=lambda num:num%2==0
print(even(5))

squre=lambda x:x**2
print(squre(9))
```

Map function is a powerfull tool used for transformation of the iterable

```python
#how to use map function with the lambda function
num=[1,2,3,4,5]
squ=lambda x:x**2
print(list(map(squ,num)))

cube=list(map(lambda x:x**3,num))
print(type(cube))
print(cube)
```

```python
help(map)
```

```python
#what to do when we have 2 data structures 
num1=[1,2,3,4,5]
num2=[3,5,7,9,11]

print(list(map(lambda x,y:x+y ,num1,num2)))
```

```python
#what if we have an convertion operation 
num3=['1','2','3','4','5']
print(list(map(int,num3)))

#we can use attribute in the operation
words=['apple','bannana','pineapple']
list(map(str.upper,words)) #do not use same name, use diff like words and str
```

```python
#map can be used with dictionries
def name_ret(dict):
    return dict['name']

"""this is a dict inside the list"""
dict=[
        {'name':'ash','age':24},
        {'name':'har','age':20},
        {'name':'sar','age':23}
    ]   


print(list(map(name_ret,dict)))


```

A filter function is  used to filter out items from iterable mainly for task such as data cleaning 

```python
#filtering even number from list by using filter function
num12=[1,2,3,4,5]
def even1(num):
    if num%2==0:
        return True

list(filter(even1,num12))
```

```python
#filtering using a lanbda function
list(filter( lambda x:x>3, num12))
```

```python
#what if we have 2 condition
list(filter( lambda x:x>3 and x%2==0, num12))
```

```python
#what if we have dictonary
def filtering_out(people):
    return people['age']>20
list(filter(filtering_out,dict))
```
