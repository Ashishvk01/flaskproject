```python
#how to convert celsius to farnheit and vice versa
def temp_conv(units,temp):
    if units=='c':
        f=float(temp)*(9/5)+32
        print(f"the temp  is {f} {units}")
    elif units=='f':
        c=float(temp-32)*(5/9)
        print(f"the temp  is {c} {units}")
    else:
        print('none')

# calling the function
temp_conv('6',25)
```

```python
#how to check strenght of the password

def ismy_passwordstrong(passwd):
    if len(passwd)<8:
        return False
    if not any(char.islower() for char in passwd):
        return False
    if not any(char.isupper() for char in passwd):
        return False
    if not any('!@#$%^&*?' for char in passwd):
        return False
    return True


print(ismy_passwordstrong('avk12345@ASDF'))
print(ismy_passwordstrong('avk12345'))

    
```

```python
#how to calculate price of the items in the cart
# {{},{},{}} This is not acceptable because sets cannot contain dictionaries, as dictionaries are mutable (changeable).
cart=[
    {'name':'apple','price':0.9,'quantity':4},
    {'name':'Bannana','price':0.3,'quantity':6},
    {'name':'mango','price':0.8,'quantity':3}
]

def cost_of_cart(cart):
    tp_price=0
    # here we do not need 2 loops as we are not going inside somthing
    for item in cart:
        tp_price+=(item['price']*item['quantity'])

    print(tp_price)

cost_of_cart(cart)
```

```python
#check weather it is palaindrome or not

def is_palindrome(word):
    word=word.lower().replace(" ","")
    if word==word[::-1]:
        return 'pallindrome'
    else:
        return 'not_pallindrome'

is_palindrome('civic')
is_palindrome('ashish')
is_palindrome("cat  tac")
```

```python
#give the factorial of a number:
def factorial(n):
    fact=1
    if n==0:
        return 1
    elif n==1:
        return 1
    else:
        while n>=2:
            fact*=n
            n=n-1
            
        return fact
    
#do factorial using recursion
def factorialrec(n):
    
    if n==0:
        return 1
    else:
        fact=n*factorialrec(n-1)   
    return fact
# calling the function:
print(factorial(8))
print(factorialrec(6))
```

```python
#count the number of words in the file

def filecount(file):
    word_count={}
    with open(file,'r') as file:  #opening the file in the read mode
        for line in file:
            words= line.split()
            for word in words:
                word=word.lower().strip(' !@#$%^&*')  #converting word into lower case and 
                                                      # spliting them based on the space and special character

                word_count[word]=word_count.get(word,0)+1 #word_count is added with one everytime it sees the same 
                                                          #in the loop
    return word_count

file='prac.txt'
filecount(file)
```

```python
#check weather the email is valid or not

import re #fucction used for patten matching
def valid_mail(mail):
    pattern=r'^[a-zA-Z0-9+-._]+@[a-zA-Z0-9.+]+\.[a-zA-Z.]+$'
    # \. is used to remove the special character status form .
    # r ' ' used when match fucnction is used
    # $ used to say This must be the END of the string.
    return re.match(pattern,mail) is not None

# is not None is used to convert the regex result into True or False by checking whether a match was found.

print(valid_mail('avk@gmail.com'))
print(valid_mail('valid_mail'))
```
