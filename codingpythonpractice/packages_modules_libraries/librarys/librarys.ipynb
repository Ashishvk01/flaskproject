```python
import json
profile={'name':'ash','age':23,'grade':'A'}

#if you get a json file you can load it as string 
jd=json.dumps(profile)
print(jd)
print(type(jd))

#if you have a string and want it load as json file
ajd=json.loads(jd)
print(ajd)
print(type(ajd))
```

```python
import csv
# help(csv)
with open('prac.csv',mode='w',newline='') as file:
    writer_csv=csv.writer(file)
    writer_csv.writerow(['name','age'])
    writer_csv.writerow(['ash',24])
    writer_csv.writerow(['har',23])
    
with open('prac.csv',mode='r') as file:
    for rows in file.readlines():
        print(rows)
```

```python
import re
pattern=r'\d+'#Without r, Python first processes the string, then regex processes it.
# In regex, \d means:Any digit from 0 to 9
#One or more times
word="this is what 123 looks like"
match=re.search(pattern,word)
print(match)
print(match.group())
```
