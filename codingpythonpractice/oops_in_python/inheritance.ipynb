Inheritance: The concept from which a class aquire the properties of another class is called inheritance.

```python
# let us consider a normal class
class car:
    def __init__(self,windows,doors,engineType):
        self.windows=windows
        self.doors=doors
        self.engineType=engineType

    def typee(self):
        print(f"the car is having {self.engineType} engine")
```

```python
car1=car(4,4,'petrol')
car1.typee()
```

```python
# now we create a class called tesla which has very similar packages to car
# so what we do we inherit it

class tesla(car):
    def __init__(self,windows,doors,engineType,is_selfDriving):
        super().__init__(windows,doors,engineType) #here do't call class method directly so do not use super
        self.is_selfDriving=is_selfDriving

    def check_for_self_driving(self):
        print(f"is this car selfdriving: {self.is_selfDriving}")

    def door_incar(self):
        print(f"the car has {self.doors} doors")
    

```

```python
tesla1=tesla(4,5,'electric',True)
tesla1.check_for_self_driving()
tesla1.door_incar()
```

```python
# multple inheritance: multiple class properties aquired bu single class

# lets use two base class Animal and pet
class Animal():
    def __init__(self,name):
        self.name=name

    def speak(self):
        print("some gybrish because this is for the practice")

class pet():
    def __init__(self,owner):
        self.owner=owner

class Dog(Animal,pet): # the animal and pet is passed as argument do not use them as function like super 
    def __init__(self,name,owner):
        Animal.__init__(self,name) #here you are calling a class method directly therefore you use self
        pet.__init__(self,owner)

    def speak(self):
        return f"hi my name is {self.name} and my owner is {self.owner}"

```

```python
dog=Dog('buddy','ashish')
print(dog.speak())
```
