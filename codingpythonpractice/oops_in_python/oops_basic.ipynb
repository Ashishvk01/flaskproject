```python
# design a bank application using class and object

class BankAccount():
    def __init__(self,name,balance):
        self.name=name
        self.balance=balance

    # how to deposit
    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} has been added to your account so your balance will be {self.balance}")

    def withdrawal(self,amount):
        if amount > self.balance:
            print("your amount is greater than your balance")

        else:
            self.balance-=amount
            print(f"{amount} has been deducted from your account so your balance will be {self.balance}")


    def balance_check(self):
        print(f"balance in your account is {self.balance}")


```

```python
acc1=BankAccount('ash',1000000)

# how to check the bank balance
acc1.balance_check()
```

```python
# how to deposit the money 
acc1.deposit(30000)
```

```python
# how to withdraw it
acc1.withdrawal(20000)
```
