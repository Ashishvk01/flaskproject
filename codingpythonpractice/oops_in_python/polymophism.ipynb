### POLYMORPHISUM
Polymorphism is the ability of the same method or function name to behave differently for different objects when inheriated. 

```python
# in order to show polymorphisum we have to do method overriding
# this is an example of hirarchial inheritance
class Animal():
    def speak():
        return "what this animal speaks"
    
class Dog(Animal):
    def speak(self):
        return "woof woof"
    
class Cat(Animal):
    def speak(self):
        return "meaw meaw"
    """TypeError: Cat.speak() takes 0 positional arguments but 1 was given, 
       if not used self this error you will get"""
    
cat=Cat()
dog=Dog()
print(cat.speak())
print(dog.speak())

# create a definitation with using the class
def speak(Animal):
    print( Animal.speak())
speak(cat)
```
